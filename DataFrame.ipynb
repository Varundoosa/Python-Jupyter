```python
import pandas as pd
foods=pd.read_csv("foods.csv")
foods.info()
```

```python
#nunique: method to find unique values in the series
# category is used to reduce memory usage; by matching particular value in a cell to a memory; ex: M/F
foods["Gender"].nunique()
foods["Gender"]=foods["Gender"].astype("category")
foods.info()
```

```python
foods["City"].nunique()
foods["City"]=foods["City"].astype("category")
foods.info()
```

```python
foods["Frequency"].nunique()
foods["Frequency"]=foods["Frequency"].astype("category")
foods["Item"].nunique()
foods["Item"]=foods["Item"].astype("category")
foods.info()
foods.sort_values(["Item","City"],ascending=[False,True])
```
